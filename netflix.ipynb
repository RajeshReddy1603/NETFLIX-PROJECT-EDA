```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'showid' and 'year' and count occurrences
id_counts = df2.groupby(["showid", "year"]).size().reset_index(name="count")

# 🔹 Select only the top 20 show IDs with the highest movie counts
top_showids = id_counts.groupby("showid")["count"].sum().nlargest(30).index
filtered_data = id_counts[id_counts["showid"].isin(top_showids)]

# 🔹 Plot using Seaborn barplot
plt.figure(figsize=(12, 6))
sns.barplot(data=filtered_data, x="showid", y="count", hue="year")

# 🔹 Set plot labels and title
plt.title("Top 30 ShowIDs with the Most Movie Releases per Year")
plt.xlabel("Show ID")
plt.ylabel("Number of Movies Released")
plt.xticks(rotation=90)  # Rotate for better readability

plt.show()
```
